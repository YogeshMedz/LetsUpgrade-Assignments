```python
from random import randint


def display_board(board):
    for e in range(1, 18):
        if (e != 3) and (e != 6) and (e != 9) and (e != 12) and (e != 15):
            print("            |            |            ")
        elif e == 3:
            print(f"     {board[1]}      |      {board[2]}     |      {board[3]}      ")
        elif e == 9:
            print(f"     {board[4]}      |      {board[5]}     |      {board[6]}      ")
        elif e == 6:
            print("======================================")
        elif e == 12:
            print("======================================")
        else:
            print(f"     {board[7]}      |      {board[8]}     |      {board[9]}      ")
```

```python
def player_marker_choose():
    choice = input('Player1 choose marker (X or O):  ')
    if(choice.upper() != 'X') and (choice.upper() != 'O'):
        print(f'Invalid Input:  {choice}')
        player_marker_choose()
    else:
        if choice.upper() == 'X':
            return 'X', 'O'
        else:
            return 'O', 'X'


def who_goes_first():
    flip = randint(0, 1)
    if flip == 0:
        return 'Player1'
    else:
        return 'Player2'
```

```python
def replay():
    # prompt to play again
    try:
        replay_reply = input('Hey!! Want to another round? (Yes / No):  ')
        if replay_reply[0].lower() == 'y':
            print('\n' * 50)
            tic_tac_toe_game()
        elif replay_reply[0].lower() == 'n':
            print('Game Over !!!')
        else:
            print('Please respond in Yes / No / Y / N')
            replay()
    except ValueError as v:
        print('Please check the value entered:  ', v)
        replay()
    except Exception as e:
        print('ERROR: please check the exception and enter the value again ', e)
        replay()
```

```python
def tic_tac_toe_game():
    test_board = ['#', '1', '2', '3', '4', '5', '6', '7', '8', '9']

    # Winning Cases
    a = test_board[1] + test_board[2] + test_board[3]
    b = test_board[4] + test_board[5] + test_board[6]
    c = test_board[7] + test_board[8] + test_board[9]
    d = test_board[1] + test_board[4] + test_board[7]
    e = test_board[2] + test_board[5] + test_board[8]
    f = test_board[3] + test_board[6] + test_board[9]
    g = test_board[1] + test_board[5] + test_board[9]
    h = test_board[3] + test_board[5] + test_board[7]

    player1_marker, player2_marker = player_marker_choose()
    print(f'Marker for Player1 is  {player1_marker}')
    print(f'Marker for Player2 is  {player2_marker}')
    print("")

    player_with_first_turn = who_goes_first()
    if player_with_first_turn == 'Player1':
        player_with_second_turn = 'Player2'
    else:
        player_with_second_turn = 'Player1'

    print(f'{player_with_first_turn} will go first')
    print("")

    display_board(test_board)
    print("")

    turn = 1
    try:
        while((a != 'XXX') and (b != 'XXX') and (c != 'XXX') and (d != 'XXX')
              and (e != 'XXX') and (f != 'XXX') and (g != 'XXX') and (h != 'XXX')
              and (a != 'OOO') and (b != 'OOO') and (c != 'OOO') and (d != 'OOO')
              and (e != 'OOO') and (f != 'OOO') and (g != 'OOO') and (h != 'OOO') and turn <= 9):
            try:
                position = int(input('Please enter the position at which you want to play: '))
            except ValueError as val:
                print('Please check the value entered:  ', val)
                continue
            if position <= 9:
                if(test_board[position] != 'X') and (test_board[position] != 'O'):
                    if(player1_marker == 'X' and player_with_first_turn == 'Player1') \
                            or (player2_marker == 'X' and player_with_first_turn == 'Player2'):
                        if turn % 2 != 0:
                            print(f'{player_with_first_turn} played  X  on position {position}')
                            test_board[position] = 'X'
                            turn = turn + 1
                        else:
                            print(f'{player_with_second_turn} played  O  on position {position}')
                            test_board[position] = 'O'
                            turn = turn + 1
                    else:
                        if turn % 2 != 0:
                            print(f'{player_with_first_turn} played  O  on position {position}')
                            test_board[position] = 'O'
                            turn = turn + 1
                        else:
                            print(f'{player_with_second_turn} played X on position {position}')
                            test_board[position] = 'X'
                            turn = turn + 1
                else:
                    print(f' Oops  {test_board[position]}  already occupied this spot, Try new position')
            else:
                print('Please enter value of Position between 1 to 9')

            # compute updated winning board
            a = test_board[1] + test_board[2] + test_board[3]
            b = test_board[4] + test_board[5] + test_board[6]
            c = test_board[7] + test_board[8] + test_board[9]
            d = test_board[1] + test_board[4] + test_board[7]
            e = test_board[2] + test_board[5] + test_board[8]
            f = test_board[3] + test_board[6] + test_board[9]
            g = test_board[1] + test_board[5] + test_board[9]
            h = test_board[3] + test_board[5] + test_board[7]

            display_board(test_board)

        else:
            if(player1_marker == 'X') and ((a == 'XXX') or (b == 'XXX') or (c == 'XXX') or (d == 'XXX')
                                           or (e == 'XXX') or (f == 'XXX') or (g == 'XXX') or (h == 'XXX')):
                print('Player1 Wins!!!, X was the symbol of the round')
            elif (player2_marker == 'X') and ((a == 'XXX') or (b == 'XXX') or (c == 'XXX') or (d == 'XXX')
                                              or (e == 'XXX') or (f == 'XXX') or (g == 'XXX') or (h == 'XXX')):
                print('Player2 Wins!!!, X was the symbol of the round')
            elif (player1_marker == 'O') and ((a == 'OOO') or (b == 'OOO') or (c == 'OOO') or (d == 'OOO')
                                              or (e == 'OOO') or (f == 'OOO') or (g == 'OOO') or (h == 'OOO')):
                print('Player1 Wins!!!, O was the symbol of the round')
            elif (player2_marker == 'O') and ((a == 'OOO') or (b == 'OOO') or (c == 'OOO') or (d == 'OOO')
                                              or (e == 'OOO') or (f == 'OOO') or (g == 'OOO') or (h == 'OOO')):
                print('Player2 Wins!!!, O was the symbol of the round')
            else:
                print('It was a Tie')

    except Exception as e:
        print('ERROR: please check the exception and enter the value again ', e)
    finally:
        replay()
```
