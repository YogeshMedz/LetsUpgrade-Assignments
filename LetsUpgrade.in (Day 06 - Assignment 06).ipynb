## Q1.
<p>Let us assume paper as the plane and a letter as a curve on the plane, then each letter divides the plane into regions. For example letters "A", "D","O", "P", "R" divide the plane into two regions so we say these letters each have one hole. Similarly, letter "B" has two holes and letters such as "C","E", "F", "K" have no holes. We say that the number of holes in the text is equal to the total number of holes in the letters of the text.</p>
<p>Write a program to determine how many holes are in a given text.</p>
<p><b>Input Format:</b></p>
<br>The only line contains a non-empty text composed only of uppercase letters of English alphabet.</br>
<p><b>Output Format:</b></p>
<br>output a single line containing the number of holes in the corresponding text.</br>
<p>EXAMPLE:</p>
<p><b>Input:</b></p>
<p>DRINKEATCODE</p>
<p><b>Output:</b></p>
<p>5</p>
<p>Explanation:
D R A O D has one hole hence total number of holes in the text is 5.</p>

```python
# List out the Alphabhets with circle in it.
s = {'A','B','D','O','P','Q','R'}
list=[]

hole = 0
li=input()

for item in li:
  if(item in s):
    if(item=='B'):
      hole=hole+2

    else:
      hole=hole+1

print(hole)
```

## <b>Map(), Filter() and Reduce() are built-in functions of Python.</b>
<p>These functions enable the functional program aspect of python.</p>
<p> In Functional Programming, the arguments passed are the only factors that decide upon the output.</p>
<p>These functions can take any other function as a parameter and can be supplied to other functions as parameters as well.</p>


```python
from IPython.display import YouTubeVideo

def display_yotube_video(url, **kwargs):
    """
    Displays a Youtube video in a Jupyter notebook.
    
    Args:
        url (string): a link to a Youtube video.
        **kwargs: further arguments for IPython.display.YouTubeVideo
    
    Returns:
        YouTubeVideo: a video that is displayed in your notebook.
    """
    id_ = url.split("=")[-1]
    return YouTubeVideo(id_, **kwargs)

display_yotube_video("https://www.youtube.com/watch?v=hUes6y2b--0", width=800, height=600)
```

## 1. Map()

<p>It applies the given function to all iterables and returns a new list.<p>
    <b>SYNTAX:</b>
<p>Map(function,iterables)<p>

```python
# without using map()
def sqr(a):
    return a*a
x= sqr(9.83)
print(x)
```

```python
# using map (funtion,iterables) - List
def sqr(a):
    return a*a
#type your code here
x=map(sqr,[1,3,8,9,10])
print(x)
print(list(x))
```

```python
# using map (funtion,iterables) - SET
def sqr(a):
    return a*a
#type your code here
x=map(sqr,[1,3,8,9,10])
print(x)
print(set(x))
```

```python
# map(function,iterables) - List
def sqr(a,b):
    return a/b
#type your code here
x= map(sqr,[1,2,3,4],[2,5,6,8,9])
print(x)
print(type(x))
print(list(x))
print(tuple(x))
print(set(x))
```

```python
# map(function,iterables) - Tuple
def sqr(a,b):
    return a/b
#type your code here
x= map(sqr,[1,2,3,4],[2,5,6,8,9])
print(x)
print(type(x))
print(tuple(x))
```

```python
# map(function,iterables) - Set
def sqr(a,b):
    return a/b
#type your code here
x= map(sqr,[1,2,3,4],[2,5,6,8])
print(x)
print(type(x))
print(set(x))
```

## 2. Lambda Functions

```python
from IPython.display import YouTubeVideo

def display_yotube_video(url, **kwargs):
    """
    Displays a Youtube video in a Jupyter notebook.
    
    Args:
        url (string): a link to a Youtube video.
        **kwargs: further arguments for IPython.display.YouTubeVideo
    
    Returns:
        YouTubeVideo: a video that is displayed in your notebook.
    """
    id_ = url.split("=")[-1]
    return YouTubeVideo(id_, **kwargs)

display_yotube_video("https://www.youtube.com/watch?v=25ovCm9jKfA", width=800, height=600)
```

<p>Everything in the world of programming requires a name, but there are few anonymous mysteris. Lambda functions or Lambda expressions are one of these kind of entities that are nameless.</p>

<p>• They are the functions that do not have any name.</p>
<p>• We also call them as anonymous or nameless functions.</p>
<p>• They are generally provided as input to other functions.</p>
<p>Lambda functions are single line functions. Therefore, we have to provide the expression along with the function itself.</p>

### Note: 'Lambda' is not a name, it is a KEYWORD.


### Use of Lambda Functions:
<p>1. One time use: They are used in places where we know that the function will be required just once. Therefore, lambda functions are also called as throw away functions</p>
<p>2. Input/Output to other functions: They also passed as inputs or returned as outputs of other higher-order functions.</p>
<p>Reduce Code Size: The body of lamda functions is written in a single line.</p>

```python
# without lambda
def sqr(a):
    return a*a
sqr(9)
```

```python
# with lambda
#type your code here
x= lambda a: a*a
x(187)
```

```python
# lambda within user defined functions
def abc(x):
    return lambda y: x+y
t=abc(4)
print(t)
print(t(5))
```
